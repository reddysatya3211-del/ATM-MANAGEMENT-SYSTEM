```python
from abc import ABC, abstractmethod

class BankAccount(ABC):
    @abstractmethod
    def transaction(self):
        pass


class Account:
    def __init__(self, holder, number, balance):
        self.name = holder
        self._acc_no = number
        self.__balance = balance

    def balance(self):
        return self.__balance

    def update_balance(self, value):
        self.__balance = value


class ATM(Account, BankAccount):
    LIMIT = 10000

    def show_bank(self):
        print("=================================")
        print("WELCOME TO SATYATEJA BANK")
        print("=================================")

    def valid(self, amt):
        return amt > 0

    def show_balance(self):
        print(f"Balance : ₹{self.balance()}")

    def add_money(self, amt):
        if self.valid(amt):
            self.update_balance(self.balance() + amt)
            print("Amount Deposited")
            self.show_balance()
        else:
            print("Invalid Amount")

    def take_money(self, amt):
        if not self.valid(amt):
            print("Invalid Amount")
        elif amt > ATM.LIMIT:
            print(f"Limit is ₹{ATM.LIMIT}")
        elif amt > self.balance():
            print("Low Balance")
        else:
            self.update_balance(self.balance() - amt)
            print("Amount Withdrawn")
            self.show_balance()

    def transaction(self):
        print("ATM Ready")


atm = ATM("Satyateja", 21911, 4000)
atm.show_bank()

pin = input("Enter PIN: ")

if pin == "2006":
    atm.transaction()
    while True:
        print("""
1. Balance
2. Deposit
3. Withdraw
4. Details
5. Exit
""")
        ch = int(input("Choice: "))

        if ch == 1:
            atm.show_balance()
        elif ch == 2:
            atm.add_money(int(input("Amount: ")))
        elif ch == 3:
            atm.take_money(int(input("Amount: ")))
        elif ch == 4:
            print("NAME :", atm.name)
            print("ACC  :", atm._acc_no)
            print(f"LIMIT: ₹{ATM.LIMIT}")
        elif ch == 5:
            print("Thank You")
            break
        else:
            print("Wrong Choice")
else:
    print("Access Denied")
```
